# **Resumo**

Este estudo investiga a relação entre o aumento dos casos de COVID-19 e a evasão escolar no ensino médio brasileiro durante a pandemia. Utilizando uma abordagem quantitativa, foram analisados dados de séries temporais provenientes de fontes oficiais, como o Our World in Data e o Censo Escolar. A metodologia empregou o modelo ARIMA para identificar padrões e prever tendências na evasão escolar correlacionadas aos picos de contaminação por COVID-19.

**Palavras-chave:** *Covid-19; Evasão escolar; Séries temporais.*


# **Introdução**

A evasão escolar é um problema presente na educação brasileira, atingindo todos os níveis de educação, o que afeta o desenvolvimento dos indivíduos e da economia do país. A permanência no ambiente escolar tende a ser um fator determinante no desenvolvimento econômico e social de um cidadão, uma vez que ela pode trazer maiores oportunidades de emprego e crescimento pessoal, sendo fundamental para a redução das desigualdades sociais e o desenvolvimento e crescimento econômico de um país. 


De acordo com Silva Filho e Araújo (2017) a “evasão e o abandono representam um processo muito complexo, dinâmico e cumulativo de saída do estudante do espaço da vida escolar”. A evasão escolar pode não estar relacionada apenas ao próprio aluno, mas também à políticas do governo e à família. Sousa et al. (2011) destacam, entre os motivos para a evasão escolar, a necessidade de ingresso no mercado de trabalho para auxiliar no sustento da família ou o próprio, o envolvimento na criminalidade e na violência, a má qualidade do ensino e conflitos no convívio familiar. 


A pandemia, iniciada em 2020, trouxe desafios para a educação ao redor do mundo. Com o fechamento das escolas e a adoção do ensino à distância e remoto, problemas como a falta de acesso à internet, ausência de equipamentos adequados e a necessidade de adaptação de toda a família ao novo modelo educacional foram fatores que dificultaram a continuidade dos estudos por parte de uma parcela de alunos. Além disso, diante dos problemas econômicos gerados pela pandemia, diversos alunos precisaram ingressar no mercado de trabalho para auxiliar no sustento familiar.


Diante deste cenário, este estudo tem como objetivo analisar a existência de uma correlação entre o aumento de casos de Covid-19 durante a pandemia e os índices de evasão escolar, em especial no ensino médio. 

# **Referencial Teórico**

As séries temporais são conjuntos de observações coletadas sequencialmente ao longo do tempo, permitindo a análise da evolução de fenômenos em diferentes períodos. No contexto educacional e de saúde pública, a modelagem de séries temporais pode ser uma ferramenta valiosa para investigar a relação entre variáveis ao longo do tempo, como a associação entre o número de casos de COVID-19 e a evasão escolar. A análise dessas séries permite identificar possíveis impactos da pandemia no abandono escolar e fornecer subsídios para políticas públicas mais eficazes.

De acordo com Morales, Rodrigues e Garcia (2024), as séries temporais são conjuntos de observações sequenciais no tempo, com intervalos regulares ou irregulares, tendo como objetivo dessas análises é identificar padrões e tendências ao longo do tempo. Segundo Bernal, Cummings e Gasparrini (2017) a quebra de segmento temporal pode ocorrer tanto por intervenções planejadas quanto por eventos inesperados que alteram significativamente o comportamento da série, como no caso da pandemia de COVID-19.

Dentro dos modelos existentes de pesquisas de séries temporais, na área da saúde o modelo mais utilizado é o ARIMA (AutoRegressive Integrated Moving Average) (Sato, 2013). Segundo Noble (2024) o modelo ARIMA é “uma técnica para análise de séries temporais e forecasting de possíveis valores futuros de uma série temporal”. 

No contexto da pandemia de COVID-19, existem diversos estudos comparativos entre os dados dos números de casos relacionados com outras variáveis. Neri e Osorio (2021) analisaram os dados de evasão escolar durante o período da pandemia. Dentre os resultados encontrados, o acesso à internet por uma grande parte dos estudantes foi fator determinante para a continuidade dos estudos de Ensino Médio e Fundamental. Ao final do estudo, os “resultados  apontam  para  uma  perda  do crescimento e da equidade na acumulação de capital humano, com efeitos negativos para o país no longo prazo”. (Neri; Osorio, 2021).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

df_covid = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv', parse_dates=['date'])

df_covid_br = df_covid[df_covid['location'] == 'Brazil'][['date', 'new_cases']].set_index('date')

df_covid_train = df_covid_br[df_covid_br.index < '2023-01-01'].copy()

df_covid_train['new_cases'] = df_covid_train['new_cases'].interpolate(method='linear')

rolling_mean = df_covid_train['new_cases'].rolling(window=7, center=True).mean()
df_covid_train['new_cases_clean'] = np.where(
    (df_covid_train['new_cases'] > 3 * rolling_mean) | (df_covid_train['new_cases'] < 0),
    rolling_mean,
    df_covid_train['new_cases']
)
df_covid_train['new_cases_clean'].fillna(method='bfill', inplace=True)

plt.figure(figsize=(14, 5))
plt.plot(df_covid_train['new_cases_clean'], label='Casos limpos')
plt.title('Novos casos de COVID-19 no Brasil (até 2022)')
plt.legend()
plt.show()

result = adfuller(df_covid_train['new_cases_clean'].dropna())
print(f'ADF_covid Statistic: {result[0]}')
print(f'p-value: {result[1]}')

df_covid_train['diff'] = df_covid_train['new_cases_clean'].diff()
fig, ax = plt.subplots(1, 2, figsize=(15, 4))
plot_acf(df_covid_train['diff'].dropna(), ax=ax[0])
plot_pacf(df_covid_train['diff'].dropna(), ax=ax[1])
plt.show()
```

# **Modelos**

Descrição metodológica para a criação, refinamento e validação dos modelos propostos.

```python
from pmdarima.arima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

serie = df_covid_train['new_cases_clean'].dropna()

# Auto-ARIMA para identificar p, d, q
auto_model = auto_arima(serie, seasonal=False, trace=True, suppress_warnings=True)
order = auto_model.order
print(f"Melhores parâmetros ARIMA: {order}")

model = ARIMA(serie, order=order)
model_fit = model.fit()

start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-12-31')
n_days = (end_date - start_date).days + 1

forecast_2023 = model_fit.forecast(steps=n_days)

forecast_index = pd.date_range(start=start_date, periods=n_days, freq='D')

plt.figure(figsize=(14, 5))
plt.plot(serie.index, serie, label='Histórico (até 2022)')
plt.plot(forecast_index, forecast_2023, label='Previsão 2023', color='red')
plt.title('Previsão de Novos Casos para o Brasil (2023)')
plt.xlabel('Data')
plt.ylabel('Novos Casos')
plt.legend()
plt.show()
```

```python
forecast_df = pd.DataFrame({
    'data': forecast_index,
    'cases': forecast_2023
})
forecast_df['Ano'] = forecast_df['data'].dt.year.astype(int)

historical_df = pd.DataFrame({
    'data': serie.index,
    'cases': serie.values
})
historical_df['Ano'] = historical_df['data'].dt.year.astype(int)

df_combined = pd.concat([historical_df[['Ano', 'cases']], forecast_df[['Ano', 'cases']]], axis=0)

df_grouped = df_combined.groupby('Ano').sum().reset_index()

print(df_grouped)
```

### Validação dos casos de Covid

```python
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error

# Supõe que df_covid já esteja carregado e filtrado para o Brasil com 'date' como índice
# E que há uma coluna 'new_cases_clean' tratada com valores ausentes resolvidos

# Substitua por sua série tratada real:
df_covid_validate = df_covid_br[df_covid_br.index < '2023-01-01'].copy()

df_covid_validate['new_cases'] = df_covid_validate['new_cases'].interpolate(method='linear')
serie = df_covid_validate['new_cases'].dropna()

# Separar treino (até 2021) e teste (2022)
serie_train = serie[serie.index < '2022-01-01']
serie_test = serie[(serie.index >= '2022-01-01') & (serie.index <= '2022-12-31')]

# Auto ARIMA
auto_model = auto_arima(serie_train, seasonal=False, trace=True, suppress_warnings=True)
order = auto_model.order
print(f"Melhores parâmetros ARIMA: {order}")

# Ajuste do modelo
model = ARIMA(serie_train, order=order)
model_fit = model.fit()

# Previsão para 2022
start_date = pd.to_datetime('2022-01-01')
end_date = pd.to_datetime('2022-12-31')
n_days = (end_date - start_date).days + 1

forecast_2022 = model_fit.forecast(steps=n_days)
forecast_index = serie_test.index

# Avaliação
rmse = mean_squared_error(serie_test, forecast_2022, squared=False)
mae = mean_absolute_error(serie_test, forecast_2022)
mape = (abs((serie_test - forecast_2022) / serie_test).dropna()).mean() * 100

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape:.2f}%")

# Plot
plt.figure(figsize=(14, 5))
plt.plot(serie_train.index, serie_train, label='Treino (até 2021)')
plt.plot(serie_test.index, serie_test, label='Real (2022)', color='black')
plt.plot(forecast_index, forecast_2022, label='Previsão (2022)', color='red')
plt.title('Validação: Previsão de Casos de COVID no Brasil (2022)')
plt.xlabel('Data')
plt.ylabel('Novos Casos')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Embora o modelo tenha capturado parcialmente o padrão geral da série, observa-se uma discrepância significativa em diversos períodos do ano, o que reforça os altos valores das métricas de erro. Essa limitação pode estar associada à alta volatilidade dos dados e à ocorrência de surtos pontuais, influenciados por variáveis exógenas não modeladas (ex.: políticas públicas, comportamento social, variantes do vírus).

Além disso, ao aplicar uma abordagem semelhante para prever os casos de abandono escolar no ensino médio utilizando uma série anual com apenas cinco observações (2018 a 2022), constatou-se uma grande limitação metodológica. A escassez de dados impossibilita a captura adequada de padrões sazonais ou tendências complexas, o que compromete a qualidade das previsões. Mesmo com essa limitação, observa-se uma tendência de crescimento na evasão escolar ao longo do período, especialmente após 2020, coincidindo com os anos críticos da pandemia.

Essa constatação sugere que, embora o modelo ARIMA possa ser útil em contextos com dados temporais mais densos (como séries diárias de COVID-19), ele apresenta limitações importantes em séries curtas e anuais, como no caso do abandono escolar. Portanto, métodos alternativos — como modelagens com variáveis exógenas (ARIMAX) ou técnicas baseadas em aprendizado de máquina — podem oferecer resultados mais robustos para esse tipo de aplicação.

# **Dicussão e Conclusão**

Este trabalho teve como objetivo analisar o impacto da pandemia de COVID-19 no abandono escolar do ensino médio brasileiro, utilizando métodos de séries temporais — especialmente o modelo ARIMA — como ferramenta de previsão. O problema de pesquisa partiu da observação do aumento expressivo nas taxas de evasão escolar durante os anos críticos da pandemia, conforme apontado por dados do INESC (2023), da Agência Gov (2024) e dos resultados do Censo Escolar 2024 divulgados pelo MEC e INEP (2025). O objetivo principal foi prever possíveis tendências futuras de abandono, buscando auxiliar na formulação de políticas públicas e estratégias de mitigação.

Para isso, o trabalho utilizou dados de duas naturezas: séries temporais diárias de novos casos de COVID-19, extraídas da base internacional Our World in Data, e dados anuais de evasão escolar obtidos em fontes governamentais e institucionais. A metodologia envolveu o pré-processamento dos dados, tratamento de valores ausentes, análise de estacionaridade com o teste ADF e identificação dos parâmetros ótimos para o modelo ARIMA por meio do auto_arima.

A avaliação da modelagem com dados de COVID-19 — em que foi possível realizar validação cruzada utilizando dados reais de 2022 — indicou desempenho razoável, com erro médio absoluto (MAE) superior a 41 mil casos e raiz do erro quadrático médio (RMSE) acima de 145 mil casos. Já no caso dos dados de abandono escolar, a aplicação do modelo ARIMA apresentou limitações significativas devido ao número reduzido de observações anuais (apenas cinco anos disponíveis). Ainda assim, foi possível visualizar uma tendência de crescimento nas taxas de evasão, especialmente após o início da pandemia, reforçando a preocupação com seus impactos educacionais de longo prazo.

Como contribuições, o trabalho demonstrou a viabilidade da utilização de modelos de séries temporais para análise de dados públicos em saúde e educação, mesmo com restrições de granularidade. Além disso, a estrutura da pipeline pode ser replicada para outros contextos onde se deseja prever comportamentos futuros com base em séries históricas.

Entretanto, algumas limitações devem ser destacadas. Primeiramente, a baixa densidade da série de evasão escolar comprometeu a qualidade das previsões. Além disso, o modelo ARIMA, por ser univariado, não incorpora variáveis externas que poderiam melhorar a acurácia, como indicadores socioeconômicos, investimentos em educação, entre outros. A instabilidade e imprevisibilidade associadas à pandemia também dificultam previsões precisas.

Para trabalhos futuros, recomenda-se o uso de modelos multivariados, como o ARIMAX ou modelos baseados em aprendizado de máquina, que permitam a inclusão de variáveis explicativas externas. Também seria relevante enriquecer a base de dados com informações regionais ou por faixa etária, o que permitiria análises mais segmentadas e políticas públicas mais direcionadas. Além disso, a combinação de abordagens quantitativas com qualitativas pode trazer novos insights sobre os fatores que impulsionam o abandono escolar em contextos de crise.
